```python
from pydantic import BaseModel, validate_call, Field
from typing import Annotated, List

@validate_call()
def add(x: int, y: int) -> float:
    return x + y
 
print(add(5,"55"))

```

```python
class User(BaseModel):
    name: Annotated[str, Field(..., title="Name of user", examples= ["Bibek Kushwaha", "Rupesh Yadav"], max_length= 50)]

    age: Annotated[int, Field(gt= 0)]

user1 = {"name": "Bibek", "age":19}
user = User.model_validate(user1)
user
```
