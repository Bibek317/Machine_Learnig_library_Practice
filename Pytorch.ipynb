```python
import torch,time,jax,jax.numpy as jnp
```

```python
# Regular Python function
def square_py(x):
    return x ** 2

# TorchScript function
@torch.jit.script
def square_ts(x):
    return x ** 2

# Create input tensor
x = torch.randn(1000, 1000)

# Measure time for regular Python function
start_time_py = time.time()
for _ in range(1000):  # Example: repeat 1000 times
    result_py = square_py(x)
elapsed_time_py = time.time() - start_time_py

# Measure time for TorchScript function
start_time_ts = time.time()
for _ in range(1000):  # Example: repeat 1000 times
    result_ts = square_ts(x)
elapsed_time_ts = time.time() - start_time_ts

# Print results
print("Elapsed time (Python function):", elapsed_time_py)
print("Elapsed time (TorchScript function):", elapsed_time_ts)
```
